```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

rd=[]
coe1=[]
for i in range (150):
    df_train, df_test = train_test_split(df_p1,test_size = 0.40, random_state= i)
    X_train = df_train.loc[:,['Danger','Predation','Gestation', 'Exposure',"TotalSleep",'omni','herbi','carni','insecti',"LifeSpan"]].values
    y_train = df_train.Dreaming.values
    X_test = df_test.loc[:,['Danger','Predation','Gestation', 'Exposure','TotalSleep','omni','herbi','carni','insecti','LifeSpan']].values
    y_test = df_test.Dreaming.values
    reg = LinearRegression()
    reg.fit(X_train, y_train)
    y_test_predicted =  reg.predict(X_test)
    Coe=r2_score(y_test, y_test_predicted)
    if Coe >= 0.5:
        print("Random state: %.2f" %i)
        print(Coe)
        rd.append(i)
        coe1.append(Coe)
rd
coe1
print("Random state:", rd)
mean_coe = sum(coe1)/len(coe1)
print("Mean of Coefficient of determination: %.2f" %mean_coe)
```
